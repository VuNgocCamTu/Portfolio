```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap='RdYlGn')

plt.title("Correlation Matrix Between Variables")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_3_manual = ['DiffPts', 'HTGD', 'HTP']

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[top_3_manual], palette="YlOrRd")

plt.title('Outlier Detection for DiffPts, HTGD, and HTP')

plt.show()
```

## Outliers Detection

The boxplots for the variables **HTP**, **DiffPts**, and **HTGD** show several outliers located outside the main distribution of the data.

### 1. HTP

- Some very high point values appear above the boxplot range.
- These values likely represent teams with exceptionally strong performance during the season.

### 2. DiffPts

- There are several large point-difference values between teams.
- These outliers indicate significant differences in team strength in certain matches.

### 3. HTGD

- Both positive and negative outliers are observed.
- This is reasonable because some teams may have very high or very low goal differences.

### Conclusion

- The detected outliers mainly reflect realistic football performance rather than data errors.
- Therefore, removing these values is not necessary at this stage.
- However, they should continue to be monitored during preprocessing and modeling to ensure they do not negatively affect model performance.
