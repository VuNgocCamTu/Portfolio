## 💡 Key Findings

* **Data quality:** No NaN values or duplicate rows in the cleaned dataset. The value `'M'` in `HM1–HM5` / `AM1–AM5` is a structured missing value (appears at the start of a season when a team lacks history). Home teams show a clear advantage: average **1.53 goals** vs away **1.13 goals**.
* **Modeling summary:** Part 3 trains and compares three models: **Logistic Regression** (baseline), **Random Forest**, and **XGBoost**. A temporal split is used (train = 2000–2014, test = 2014–2018). Key features include `DiffPts`, `DiffFormPts`, `HTGD`, `ATGD` and one-hot encoded recent results `HMx`, `AMx`.
* **Evaluation:** Models are compared using **Accuracy, Precision, Recall, F1-Score, ROC-AUC** and **Confusion Matrix**. A majority-class baseline is established; models that beat the baseline demonstrate learned predictive signal.
* **Top features:** `DiffPts` and `DiffFormPts` are consistently among the most important features, indicating team strength and recent form strongly influence match outcomes.

## 🧾 Detailed Report — 3 Parts

### 1. Data Understanding
- Objective: understand dataset structure and quality before processing.
- Actions: loaded `final_dataset.csv`, inspected shape, sample rows, `df.info()`, `df.describe()`, missing values, and the `'M'` values in `HMx` / `AMx`.
- Conclusion: dataset contains **6,840 matches** and **39 columns**; data are pre-match statistics without large NaN issues. `Date` should be converted to datetime and recent-result columns need encoding.

### 2. Data Cleaning
- Objective: produce a clean dataset without leakage, ready for modeling.
- Actions:
	- converted `Date` to datetime and sorted chronologically
	- removed columns that leak information or are redundant:
		- `FTHG`, `FTAG` (match results — data leakage)
		- `HTGS`, `ATGS`, `HTGC`, `ATGC` (aggregated into `HTGD`, `ATGD`; potential multicollinearity)
		- `HM4`, `HM5`, `AM4`, `AM5` (4–5 match history has weaker signal than last 3 matches)
		- `HomeTeam`, `AwayTeam`, `HTFormPtsStr`, `ATFormPtsStr` (identifiers or redundant representations)
	- scaled cumulative variables by `MW` to produce per-match averages for `HTGD`, `ATGD`, `DiffPts`, `DiffFormPts`, `HTP`, `ATP`
	- dropped `MW` after scaling
	- one-hot encoded `HM1–HM3`, `AM1–AM3`, keeping `M` as a separate category (not treated as a loss)
- Result: saved `cleaned_data.csv` containing the cleaned features used for modeling.

### 3. Model Training
- Objective: train and compare models to predict `FTR` using a temporal split.
- Actions:
	- loaded `cleaned_data.csv`, converted `Date` to datetime and sorted by date
	- selected target `FTR` and a concrete feature set:
		- `DiffPts`, `DiffFormPts`, `HTGD`, `ATGD`
		- `HTWinStreak3`, `HTWinStreak5`, `HTLossStreak3`, `HTLossStreak5`
		- `ATWinStreak3`, `ATWinStreak5`, `ATLossStreak3`, `ATLossStreak5`
		- one-hot columns: `HM1_*`, `HM2_*`, `HM3_*`, `AM1_*`, `AM2_*`, `AM3_*`
	- split data temporally: train = first 80% of matches, test = last 20%
	- standardized features using `StandardScaler` fit on train and applied to test
	- established a majority-class baseline from the train set
	- trained three models: `LogisticRegression`, `RandomForestClassifier`, `XGBClassifier`
	- evaluated with `Accuracy`, `Precision`, `Recall`, `F1-Score`, `ROC-AUC`
	- plotted confusion matrices and ROC curves; computed feature importances
- Conclusion: models trained on the cleaned, temporally-split data; `DiffPts` and `DiffFormPts` are key predictors.
